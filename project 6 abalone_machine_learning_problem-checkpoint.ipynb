```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from scipy.stats import zscore
import warnings
warnings.filterwarnings('ignore')

df_abalone=pd.read_csv("abalone.csv",sep=",",na_values=".")
df_abalone
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_abalone['Sex']=le.fit_transform(df_abalone['Sex'])
```

```python
df_abalone
```

```python
x=df_abalone.drop(columns=["Rings"])
y=df_abalone[["Rings"]]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state = 45,test_size=0.20)
lm=LinearRegression()
lm.fit(x_train,y_train)
pred = lm.predict(x_test)
print(pred)
```

```python
print("RMSE is: ",np.sqrt(mean_squared_error(y_test,pred)))
print("r2_score is: ",r2_score(y_test,pred) )
```
